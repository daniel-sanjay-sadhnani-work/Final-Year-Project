```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Excel file
file_path = 'Cleansed.xlsx'  # Update the file path accordingly
df = pd.read_excel(file_path)

# Extract the 'Skill 3' column and remove NaN values
skills = df['Skill 3'].dropna()

# Convert text data to numerical form using term frequency (CountVectorizer)
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(skills)

# Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.5, min_samples=2, metric='cosine')
clusters = dbscan.fit_predict(X.toarray())

# Create a new DataFrame with skills and their corresponding clusters
clustered_df = pd.DataFrame({'Skill 3': skills, 'Cluster': clusters})

# Display the first few rows of the DataFrame with cluster labels
print(clustered_df.head())

# Plot the cluster distribution using a scatter plot
plt.figure(figsize=(12, 8))
sns.scatterplot(x=range(len(clustered_df)), y='Skill 3', hue='Cluster', data=clustered_df, palette='tab10')
plt.xlabel('Index')
plt.ylabel('Skill 3')
plt.title('Scatter Plot of Skills Clustering')
plt.legend(title='Cluster')
plt.show()
```

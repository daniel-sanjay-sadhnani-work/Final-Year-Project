```python
import pandas as pd
```

```python
df = pd.read_excel("Student Survey.xlsx")
df
```

```python
df = df.drop(columns = "Timestamp")
```

```python
df = df.rename(columns = {'What is your gender?':'Gender', 'How old are you this year?' : 'Age','Which tertiary education institution are you in?':'Tertiary Education','Of what ethnicity are you':'Ethnicity','Of what religion are you?':'Religion','Do your parents work?':'Parents Work','How many siblings do you have':'Siblings','What type of housing do you live in?':'Housing','What is your current cGPA? Please answer  in 2 decimal points(e.g., 3.30)':'Current GPA','What is your desired cGPA?  Please answer  in 2 decimal points(e.g., 3.30)':'Desired GPA','On average, how many times are you absent for class in a semester':'Absences','Do you play video games?':'Video Games','How many Hours  do you game per week?':'Game Hours','What Game Genre Do You Mostly Play?':'Game Genre','Do You Play Games With Friends?':'Game With Friends','During what hours do you usually start playing video games? Please answer in the 24 hour time format (e.g., 0000, 1800, 2200 etc)':'Game Timing','Do your Parent(s) Stop You From Gaming?':'Gaming Stopped','What Is The Main Reason You Started Gaming':'Main Reason Started Gaming','Are you working as a part-timer as of right now?':'Part-Time','How Many Hours Do You Work A Week?':'Part-Time Hours','What Job Sector Do You Work In?':'Job Sector','Which Days Of The Week Do You Work?':'Days Of Week Worked','Do you participate in any Co-Curricular Activities (CCAs) at your school?':'CCAs Participated','How many Co-Curricular Activities (CCAs) are you currently participating in?':'No. Of CCAs Participated','Please specify the name(s) of the CCA(s) you are involved in.':'Names Of CCAs','What role do you hold in your CCA(s)? ':'CCA Role','How many hours per week do you dedicate to your CCA(s)? ':'CCA Hours','On a scale of 1 to 10, how satisfied are you with your overall school experience? (1 being very unsatisfied, 10 being very satisfied)':'School Satisfaction Level','Which specific area of your school experience do you feel most satisfied with?':'Satisfied Area','Which specific area of your school experience are you not satisfied with? ':'Unsatisfied Area','\nHow do you rate the support provided by your school in achieving your personal and academic goals? (1 being very poor, 10 being excellent)':'Support Level','Do you feel that your school fosters a sense of community and belonging among its students? ':'Sense Of Belonging','Do have your future career planned already?':'Future Planned','Write down at least 3 career options after graduation (e.g. 1. teacher 2. chef  3. doctor)':'3 career options after graduation','Write down if there is any 3 career option after graduation that you would be interested to do  (e.g. 1. teacher 2. chef  3. doctor)':'3 career option interested to do','With the career options that you have stated, briefly give the skills that is required for the options.(e.g. 1. Patience 2. Cooking 3. Precision)':'Skills Required'})
df
```

```python
df = df.loc[df['Tertiary Education'].str.contains('Polytechnic', case=False, na=False)]
df
```

```python
print(df.columns)
```

```python
df['Siblings'] = df['Siblings'].replace({'5 siblings': '5'})

# Print the unique values in the 'Siblings' column to verify the replacement
print(df.iloc[32])
```

```python
df
```

```python
df['Current GPA'] = pd.to_numeric(df['Current GPA'], errors='coerce')

df = df.dropna(subset=['Current GPA'])
df
```

```python
df['Desired GPA'] = pd.to_numeric(df['Desired GPA'], errors='coerce')

df = df.dropna(subset=['Desired GPA'])
df
```

```python
df = df[(df['Current GPA'] >= 0) & (df['Current GPA'] <= 4)]
df
```

```python
df = df[(df['Desired GPA'] >= 0) & (df['Desired GPA'] <= 4)]
df
```

```python
df.columns
```

```python
df['Absences'] = pd.to_numeric(df['Absences'], errors='coerce')

df = df.dropna(subset=['Absences'])
df
```

```python
columns_to_replace = ['Game Hours', 'Game Genre',
       'Game With Friends', 'Game Timing', 'Gaming Stopped',
       'Main Reason Started Gaming']
df[columns_to_replace] = df[columns_to_replace].replace(['NaN', None], "Don't Play Video Games")
df
```

```python
print(df['Game Hours'])
```

```python
columns_to_replace = ['Part-Time Hours',
       'Job Sector', 'Days Of Week Worked',]
df[columns_to_replace] = df[columns_to_replace].replace(['NaN', None], "Don't Work")
df
```

```python
print(df['Part-Time Hours'])
```

```python
df = df.rename(columns={'Please specify the name(s) of the CCA(s) you are involved in. ': 'CCA Names'})
```

```python
df.columns
```

```python
columns_to_replace = ['No. Of CCAs Participated', 'CCA Names', 'CCA Role', 'CCA Hours']
df[columns_to_replace] = df[columns_to_replace].replace([None], "Don't Have CCA")
df
```

```python
df['CCA Hours'] = df['CCA Hours'].fillna("Don't Have CCA")
df
```

```python
print(df['CCA Hours'])
```

```python
df.to_excel('cleansed_data.xlsx', index=False)
```
